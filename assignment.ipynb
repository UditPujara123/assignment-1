```python
def month(data):
    if 0<data<=12:
        return 1
    elif 12<data<=24:
        return 2
    elif 24<data<=36:
        return 3
    elif 36<data<=48:
        return 4
    elif 48<data<60:
         return 5
    elif 60<=data<=72:
        return 6
    elif 72<data<=90:
        return 7
    elif 90<data<=102:
        return 8
    else:
        return 9
        
```

```python
loans["Months_loan_taken_for"]=loans["Months_loan_taken_for"].apply(month)
```

```python
x=loans.iloc[:,:-1]
y=loans.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
x=sc.fit_transform(x)
```

```python
from minisom import MiniSom
```

```python
som=MiniSom(10,10,input_len=23)
```

```python
som.random_weights_init(x)
```

```python
som.train_random(data=x,num_iteration=100)
```

```python
from pylab import bone,pcolor,colorbar,plot,show
```

```python
plt.figure(figsize=(12,6))
bone()
pcolor(som.distance_map().T)
colorbar()
markers=["o","s"]
color=["r","g"]
for i,X in enumerate(x):
    w=som.winner(X)
    plot((w[0]+0.5),(w[1]+0.5),markers[y[i]],markeredgecolor=color[y[i]],markeredgewidth=2)
```

```python
mappings=som.win_map(x)
frauds=np.concatenate((mappings[(4,4)],mappings[(4,5)]),axis=0)
```
